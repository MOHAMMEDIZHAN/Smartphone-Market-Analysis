```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Sales (1).csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# Drop the camera column
df.drop('Camera', axis=1, inplace=True)
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.describe()
```

```python
# count duplicates in models and Mobile columns
df.duplicated(subset=['Models', 'Mobile']).sum()
```

```python
# drop duplicates in the models and mobile column
df.drop_duplicates(subset=['Models', 'Mobile'], inplace=True)
```

```python
df.head(10)
```

```python
# Remove "GB", "MB" from Memory and Storage
df['Memory'] = df['Memory'].str.replace('GB', '')
df['Memory'] = df['Memory'].str.replace('MB', '')
```

```python
df['Storage'] = df['Storage'].str.replace('GB', '')
df['Storage'] = df['Storage'].str.replace('MB', '')
```

```python
df.head(10)
```

```python
# Change the datatype of Memory and storage columns to float
df['Memory'] = pd.to_numeric(df['Memory'], errors='coerce')
df['Storage'] = pd.to_numeric(df['Storage'], errors='coerce')
```

```python
df['Storage'] = pd.to_numeric(df['Storage'], errors='coerce')
median_storage = df['Storage'].median()
if pd.isna(median_storage):
    median_storage = 0
df['Storage'] = df['Storage'].fillna(median_storage)

import numpy as np
if np.isinf(df['Storage']).any():
    df['Storage'] = df['Storage'].replace([np.inf, -np.inf], median_storage)

df['Storage'] = df['Storage'].astype(int)
```

```python
df.info()
```

```python
df['Memory'].fillna(df['Memory'].median(), inplace=True)
df['Storage'].fillna(df['Storage'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Rating'].fillna(df['Rating'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(10)
```

```python
# Rename the column Memory to "RAM" and storage to "ROM"
df.rename(columns={'Memory': 'RAM', 'Storage': 'ROM'}, inplace=True)
```

```python
df.head()
```

```python
# Limit the discount percentage values to only one decimal number
df['discount percentage'] = df['discount percentage'].apply(lambda x: round(x, 1))
```

```python
df.head()
```

```python
df.drop('Mobile', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
# Price segment
def price_category(price):
    if price < 10000:
        return "Budget"
    elif price < 30000:
        return "Mid-Range"
    else:
        return "Premium"

df['Price_Category'] = df['Selling Price'].apply(price_category)
```

```python
df.head()
```

```python
df['Savings'] = df['Original Price'] - df['Selling Price']
```

```python
df.head()
```

```python
df['Has_Discount'] = df['discount percentage'].apply(
    lambda x: 1 if x > 0 else 0
)
```

```python
df['Price_per_GB'] = df['Selling Price'] / df['ROM']
```

```python
df.head()
```

```python
df['Price_per_GB'] = df['Price_per_GB'].apply(lambda x: round(x, 1))
```

```python
df.head()
```

```python
# Ram Category
def ram_category(ram):
    if ram <= 4:
        return "Low"
    elif ram <= 6:
        return "Medium"
    else:
        return "High"

df['RAM_Category'] = df['RAM'].apply(ram_category)
```

```python
# Rating Category
def rating_category(r):
    if r >= 4.5:
        return "Excellent"
    elif r >= 4.0:
        return "Good"
    else:
        return "Average"

df['Rating_Category'] = df['Rating'].apply(rating_category)
```

```python
# Unique ID for each Product
df['Product_ID'] = range(1, len(df) + 1)
```

```python
df.head(10)
```

```python
# Ordering the Columns
df = df[
    [
        'Product_ID',
        'Brands',
        'Models',
        'Colors',
        'RAM',
        'ROM',
        'Rating',
        'Selling Price',
        'Original Price',
        'discount percentage',
        'Price_Category',
        'Savings',
        'Has_Discount',
        'Price_per_GB',
        'RAM_Category',
        'Rating_Category'
    ]
]
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.to_csv(
    "final_smartphone_sales_data.csv",
    index=False
)
```
